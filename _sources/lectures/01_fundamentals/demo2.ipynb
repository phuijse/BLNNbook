# demo bokeh

```python
import numpy as np
import scipy.stats
import ipywidgets as widgets
#from bokeh.layouts import column, row
#from bokeh.models import CustomJS, ColumnDataSource, Slider
from bokeh.io import show, output_notebook, push_notebook
from bokeh.plotting import figure

import numpy as np
x = ['0', '1']
p = np.linspace(0, 1, num=5)
pmf = np.array([[scipy.stats.bernoulli(p=p_).pmf(k) for k in [0, 1]] for p_ in p])

output_notebook()
fig = figure(title='PMF of the Bernoulli distribution', x_range=x, 
             plot_height=300, plot_width=350, x_axis_label='x', y_axis_label='PMF', y_range=(0, 1.05))
r = fig.vbar(x=x, top=pmf[1, :], width=0.6)
show(fig, notebook_handle=True)


@widgets.interact(p_=widgets.SelectionSlider(options=p))
def update(p_):
    print(pmf[int(p_*5), :])
    r.data_source.data['top'] = pmf[int(p_*5), :]
    push_notebook()
```
